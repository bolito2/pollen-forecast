Nos centramos en cupresaceas por ahora

```python
f = h5py.File('pollen.h5', 'r')
madrid_subiza = np.array(f['madrid_subiza'])

cupresaceas = madrid_subiza.T[7]
print(cupresaceas.shape)
```

Vamos a normalizar los datos

```python
#cupresaceas = (cupresaceas - cupresaceas.mean())/cupresaceas.std()
print(cupresaceas[:10])
```

Separamos en años y los permutamos. De esta forma cada set tendrá una distribución similar al resto

```python
data_set = cupresaceas[len(cupresaceas)%365:]
print(data_set.shape)
data_set = tf.reshape(data_set, [-1 , 365])

data_set = tf.random.shuffle(data_set)

print(data_set.shape)
```

PRUEBA - Normalizar cada año individualmente

```python
data_set = np.array(data_set)

index_data = np.zeros((41, 365, 2))

for i in range(data_set.shape[0]):
    index_data[i, :, 0] = range(365)
    index_data[i, :, 1] = (data_set[i] - data_set[i].mean())/data_set[i].std()
    
print(index_data.shape)
```

Separamos los años en train, dev y test sets. Enseñamos los tres años del test_set, cada uno de un color

```python
def train_dev_test_split(data, train_rate, dev_rate):
    train_set = data[:int(data.shape[0]*train_rate), :, :]
    dev_set = data[int(data.shape[0]*train_rate):int(data.shape[0]*(train_rate + dev_rate)), :, :]
    test_set = data[int(data.shape[0]*(train_rate + dev_rate)):, :, :]
    
    return train_set, dev_set, test_set
```

```python
train_set, dev_set, test_set = train_dev_test_split(index_data, 0.8, 0.15)

print('train_set.shape: ', train_set.shape)
print('dev_set.shape: ', dev_set.shape)
print('test_set.shape: ', test_set.shape)

for i in range(3):
    plt.plot(test_set[i, :, 1])
    
plt.show()
```

Ahora acoplamos los años de cada set para conseguir una única secuencia de datos

```python
train_set = tf.reshape(train_set, [-1, 2])
dev_set = tf.reshape(dev_set, [-1, 2])
test_set = tf.reshape(test_set, [-1, 2])

plt.plot(test_set[:, 1])
plt.show()
```

Vamos a procesarlos en secuencias de 10 donde predecimos el último dato de la secuencia,

```python
def sliding_windows(data, size):
    if data.shape[0] % size != 0:
        data = data[:data.shape[0] - (data.shape[0] % size), :]

    windows = np.zeros((data.shape[0] - size, size, 2))
    
    for i in range(data.shape[0] - size):
        windows[i, :, 0] = data[i:i+size, 0]
        windows[i, :, 1] = data[i:i+size, 1]
        
    return windows
```

```python
window_size=10

train_windows = sliding_windows(train_set, window_size)
dev_windows = sliding_windows(dev_set, window_size)
test_windows = sliding_windows(test_set, window_size)

print(test_windows.shape)
print(test_windows[0])
```

```python
X_train = train_windows[:, :-1, :]
Y_train = train_windows[:, -1:, 1]

X_dev = dev_windows[:, :-1, :]
Y_dev = dev_windows[:, -1:, 1]

X_test = test_windows[:, :-1, :]
Y_test = test_windows[:, -1:, 1]

Y_dev = np.ones((2180, 1))
```

```python
print('X_train.shape: ', X_train.shape)
print('Y_train.shape: ', Y_train.shape)

print('X_dev.shape: ', X_dev.shape)
print('Y_dev.shape: ', Y_dev.shape)

print('X_test.shape: ', X_test.shape)
print('Y_test.shape: ', Y_test.shape)

plt.plot(Y_test)
plt.show()
```

```python
X_input = Input(shape=(window_size-1,2))

X = LSTM(10)(X_input[:, -1:])

X_out = Dense(units = 1)(X)

print(X_out.shape)
```

```python
model = Model(inputs=X_input, outputs=X_out)
model.summary()
```

```python
opt = Adam(learning_rate=0.00005, beta_1= 0.9, beta_2=0.99, epsilon=1e-7, clipnorm=10)

model.compile(loss='mse', optimizer=opt)
fitting = model.fit(X_train, Y_train, batch_size=128, epochs=20, validation_data=(X_dev, Y_dev))

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(fitting.history['loss'], color='r')
ax.plot(fitting.history['val_loss'], color='b')

plt.show()
```

```python
test_scores = model.evaluate(X_test, Y_test, batch_size=128)
```

```python
sample_start = 450
sample_end = 650

fig = plt.figure()

pred = model(X_test[sample_start:sample_end, :, :])


ax = fig.add_axes([0,0,1,1])
ax.plot(range(sample_start, sample_end), Y_test[sample_start:sample_end, 0], color='b')

ax.plot(range(sample_start, sample_end), pred[:, 0], color='r')

plt.show()
```
